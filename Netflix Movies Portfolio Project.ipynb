```python
#import libraries 
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

#Read in the data
df = pd.read_csv(r'C:\Users\yashv\Downloads\movies.csv')
```

```python
pd.set_option('display.max_rows', 20)
```

```python
#Looking at the data
df.head()
```

```python
df = df.dropna();
```

```python
df.head()
```

```python
# Checking for missing data
df.isnull().sum() 
```

```python
# Check Outliers

df.boxplot(column=['gross'])
plt.show()
```

```python
# Data types for our columns
df.dtypes
```

```python
# Change data type of columns
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')

df['runtime'] = df['runtime'].astype('int64')
```

```python
df
```

```python
# Create correct year column
df['yearcorrect'] = df['released'].astype(str).str.split().str[2]

df
```

```python
df = df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
# Drop duplicates
df.drop_duplicates()
```

```python
# Scatter plot with budget vs gross
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget of Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# Plot budget vs gross using seaborn
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"violet"}, line_kws={"color":"teal"})
plt.show()
```

```python
sns.regplot(x="score", y="gross", data=df, scatter_kws={"color":"orange"}, line_kws={"color":"blue"})
plt.show()
```

```python
df.corr(numeric_only=True)
```

```python
# There is high correlation between budget and gross
```

```python
correlation_matrix = df.corr(numeric_only=True, method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix For Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized = df.copy()

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
correlation_matrix = df_numerized.corr(numeric_only=True, method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix For Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
# High correlation
high_corr = sorted_pairs[(sorted_pairs)>0.5]
high_corr
```

```python
# Votes and budget have the highest correlation to the gross earnings
```
